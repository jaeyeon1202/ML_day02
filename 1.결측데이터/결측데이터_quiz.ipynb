### 결측데이터 치환

```python
# 데이터셋 다시 불러오기
df = pd.read_csv('data/bicycle.csv')
df.head()
```

```python
# 결측치 확인
df.isnull().sum()
```

```python
# 모든 결측치를 0으로 치환, 자기 자신은 변경되지 않습니다.
df.fillna(value=0)
```

```python
# 모든 결측치를 이전 값으로 치환
df.fillna(method='pad')
```

```python
# 이용거리 결측치 열을 이용거리 평균으로 치환, 자기자신을 변경
df['이용거리'].fillna(df['이용거리'].mean(), inplace=True)
df.head()
```

```python
# 모든 결측치를 각 열의 평균으로 처리
df.fillna(df.mean(), inplace=True)
df.head()
```

```python
# 데이터셋 다시 불러오기
df = pd.read_csv('data/bicycle.csv')
df.head()
```

```python
# 이용거리를 최빈값으로 치환하세요
print('데이터 개수\n', df['이용거리'].value_counts())
print('==================================')
print('최빈값: ',df['이용거리'].mode())
df['이용거리]'.fillna(df['이용거리'].mode()[0], inplace=True)
```

```python
# 결측 개수 확인. 이용거리 결측값 없어야 된다
print('데이터 행, 열: ', df.shape)
print('=============================')
df.isnull().sum()
```
