### 1. replace
* replace( to_replace=비교값, value=치환값)
* nan값은 np.nan으로 비교

```python
import pandas as pd
```

```python
#데이터셋 로드
df = pd.read_csv('data/nan_test.csv')
df
```

```python
import numpy as np
df['nan_1'] == np.nan
```

```python
df.replace(to_replace=np.nan, value=3000)
```

```python
df = df.replace(to_replace = {'nan_2' : np.nan}, value = 'nan2변경')
df
#값 변경할때 inplace하거나 저장해서 출력
```

```python
df.replace( {'nan_1' : np.nan},12345, inplace = True)
df
```
